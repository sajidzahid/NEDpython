```python
df ["Fare_class"] = ""
for i in df.index:
  if df.loc[i, "Fare"] > 70:
    df.loc[i,"Fare_class"]= "elite"
  elif df.loc[i, "Fare"] > 30 and df.loc[i,"Fare"] < 69:
    df.loc[i,"Fare_class"]= "Avergae"
  else:
    df.loc[i,"Fare_class"]= "Basic"
```
