```python
# prompt: 2/3 - 852/65 can we generate code to solve this type of data

def calculate(expression):
  """
  Evaluates a mathematical expression string.

  Args:
    expression: A string representing a mathematical expression.

  Returns:
    The result of the evaluation, or an error message if the expression is invalid.
  """
  try:
    result = eval(expression)
    return result
  except ZeroDivisionError:
    return "Error: Division by zero"
  except (SyntaxError, NameError, TypeError):
    return "Invalid input. Please enter a valid mathematical expression."

# Example usage
expression = "2/3 - 852/65"
result = calculate(expression)
print(result)


expression = "10 + 5"
result = calculate(expression)
print(result)

expression = "10 / 0"
result = calculate(expression)
print(result)

expression = "invalid expression"
result = calculate(expression)
result
```
