# LIST COMPREHENSIONS  USED BRACKETS [ ]

```python
numbers = []
for i in range(0,100):
  numbers.append(i)
  print(numbers)
```

```python
import time
start = time.time()
numbers = []
for i in range(0,100):
  numbers.append(i)
  end = time.time()
  print(end-start)
  print(numbers)
```

```python
numbers = [x**2 for x in range(0,100)]
print(numbers)
```

```python
numbers = [x**2 for x in range(1,100)]
start = time.time()
print(numbers)
end = time.time()
print(end-start)
```

```python
numbers = [i for i in range(0,1000) if i%2!=0] # ODD Numbers
print(numbers)
```

```python
numbers = [i for i in range(0,1000) if i%2 == 0] # Even Numbers
print(numbers)
```

# GENERATOR EXPRESSION   USE ROUND BRACKETS

```python
numbers = (i for i in range(0,10) if i%2 == 0) # This is called Generator Expressions. Always between round brackets.
print(numbers)
print(type(numbers))
```

```python
for i in numbers:
  print(i)
```
