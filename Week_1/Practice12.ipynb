```python
# prompt: https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv
# now perform EDA on this data according to their data types.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Download the Titanic dataset
!wget https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv

# Load the dataset into a pandas DataFrame
titanic_data = pd.read_csv('titanic.csv')


# Perform EDA based on data types
def eda_by_dtype(df):
    for col in df.columns:
        print(f"Analysis for column: {col}")
        print(f"Data Type: {df[col].dtype}")
        if pd.api.types.is_numeric_dtype(df[col]):
            print(f"Mean: {df[col].mean()}")
            print(f"Median: {df[col].median()}")
            print(f"Standard Deviation: {df[col].std()}")
            print(f"Minimum: {df[col].min()}")
            print(f"Maximum: {df[col].max()}")
            plt.figure()
            sns.histplot(df[col])
            plt.title(f"Distribution of {col}")
            plt.show()

            plt.figure()
            sns.boxplot(x = df[col])
            plt.title(f"Box Plot of {col}")
            plt.show()

        elif pd.api.types.is_categorical_dtype(df[col]) or pd.api.types.is_object_dtype(df[col]):
            print(f"Unique Values: {df[col].unique()}")
            print(f"Value Counts:\n{df[col].value_counts()}")
            plt.figure()
            df[col].value_counts().plot(kind='bar')
            plt.title(f"Value Counts of {col}")
            plt.show()

        elif pd.api.types.is_datetime64_any_dtype(df[col]):
          print(f"Earliest Date: {df[col].min()}")
          print(f"Latest Date: {df[col].max()}")

        print("-" * 50)


eda_by_dtype(titanic_data)
```
